```python
from matplotlib import pyplot as plt, font_manager
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\STSONG.TTF', size=18)
a = ['流浪地球', '疯狂外星人', '飞驰认识','大黄蜂','熊出没','喜剧之王']
b = [38.13, 19.85, 14.89, 11.36, 47.5, 5.93]
plt.figure(figsize=(20,10), dpi=80)
# 条形图参数 1:个数    2:高度    3:宽度
rects = plt.bar(range(len(a)), b, width=0.3, color ='red')
plt.xticks(range(len(a)),a,fontproperties=my_font)
# 标注显示
for rect in rects:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() /2, height+0.3,str(height), ha="center")
plt.show()
```

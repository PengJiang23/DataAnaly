```python
from matplotlib import pyplot as plt, font_manager
import random
# 导入字体配置
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\STSONG.TTF', size=18)

x = range(2,26,2)
y = [random.randint(15,30) for i in x]
plt.figure(figsize=(10,5),dpi=80)
x_ticks_label = ['{}:00'.format(i) for i in x]
plt.xticks(x,x_ticks_label,rotation = 45)
y_ticks_label = ['{}°C'.format(i) for i in range(min(y), max(y)+1)]
"""
ticks参数 1：刻度个数  2：标签值   3：角度
"""
plt.yticks(range(min(y), max(y)+1), y_ticks_label, rotation = 0)
plt.plot(x,y, color='red', alpha = 0.5, linestyle = '-', linewidth = '5', marker = 'o')
# 标签
plt.xlabel('时间', rotation=45, fontproperties = my_font)
plt.ylabel('温度', fontproperties = my_font)
plt.title('一天随机温度变化表', color = 'green', fontproperties = my_font)

plt.savefig('./1.svg')
plt.show()
```

```python
from matplotlib import pyplot as plt, font_manager
import random

# 导入字体配置
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\STSONG.TTF', size=18)

x = range(2, 26, 2)
y = [random.randint(15, 30) for i in x]
z = [random.randint(15, 30) for i in x]
plt.figure(figsize=(10, 5), dpi=80)

x_ticks_label = ['{}:00'.format(i) for i in x]
plt.xticks(x, x_ticks_label, rotation=45)
y_ticks_label = ['{}°C'.format(i) for i in range(min(y), max(y) + 1)]
"""
ticks参数 1：刻度范围  2：标签值   3：角度
"""
plt.yticks(range(min(y), max(y) + 1), y_ticks_label, rotation=0)
plt.plot(x,y,color = 'red', label = 'a1')
plt.plot(x,z,color = 'blue', label = 'a2')
# 标签
plt.xlabel('时间', rotation=45, fontproperties=my_font)
plt.ylabel('温度', fontproperties=my_font)
plt.title('随机温度变化表', color='green', fontproperties=my_font)
# 图例
plt.grid(alpha = 0.6)
plt.legend(prop=my_font, loc = 'upper right')

plt.show()
```

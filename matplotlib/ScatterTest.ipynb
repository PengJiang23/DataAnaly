```python
from matplotlib import pyplot as plt, font_manager
import random
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\STSONG.TTF', size=18)
x = range(1,32)
y = [random.randint(10,20) for i in x]
x_ticks_label = ['3月{}日'.format(i) for i in x]
# 刻度值和标签值一一对应 否则ValueERROR
plt.xticks(x[::3],x_ticks_label[::3], fontproperties = my_font, rotation = 45)
plt.xlabel('时间', fontproperties = my_font)
plt.ylabel('温度', fontproperties = my_font)
plt.scatter(x,y, label = '温度点')
plt.grid(alpha = 0.3)
plt.legend(prop = my_font, loc='upper left')
plt.show()
```

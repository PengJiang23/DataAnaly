```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt, font_manager

# 星巴克咖啡统计
file_path = './directory.csv'
df = pd.read_csv(file_path)
print(df.info())
print(df.head(1))
dc = df.groupby(by='Country')
# for i,j in dc:
#     print(i)
#     print('-'*20)
#     print(j,type(j))
#     print('-'*20)
china_data = df[df['Country'] == 'CN']
print(china_data.groupby(by='State/Province').count()['Brand'])
```

```python
# 店铺总数前十的国家
data = df.groupby('Country').count()['Brand'].sort_values(ascending=False)[:10]
_x = data.index
_y = data.values

plt.bar(range(len(_x)),_y,width=0.5)
plt.xticks(range(len(_x)),_x)
plt.show()
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt, font_manager
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\STSONG.TTF', size=14)
# 星巴克咖啡统计
file_path = './directory.csv'
df = pd.read_csv(file_path)
# 每个中国每个城市的店铺数量
data = df[df['Country']=='CN']
d1 = data.groupby(by='City').count()['Brand'].sort_values(ascending=False)[:10]
_x = d1.index
_y = d1.values
plt.bar(range(len(_x)),_y,width=0.3,color = 'orange')
plt.xticks(range(len(_x)),_x,fontproperties=my_font,rotation=30)
plt.show()
```

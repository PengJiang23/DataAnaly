```python
import pandas as pd
from matplotlib import pyplot as plt
# 书本统计
file_path = './books.csv'
df = pd.read_csv(file_path)
print(df.info())
# 年份为空的数据需要舍弃
df1 = df[pd.notnull(df['original_publication_year'])]
df2 = df1.groupby(by='original_publication_year').count()['title']
_x = df2.index
_y = df2.values
plt.plot(range(len(_x)),_y)
plt.show()
```
